```python
class Node:
    def __init__(self,data,left,right):
        self.data=data
        self.left=None
        self.right=None
```

```python
class BST:
    def __init__(self):
        self.root=None
    def insert(self,data):
        if self.root is None:
            self.root=Node(data)
            return
        if self.root.data< data: # root is found
            if self.right:
                self.right.insert(data)
            else:
                self.right=BST(data)
        else:
            if self.left:
                self.left.insert(data)
            else:
                self.left=BST(data)
    def get_nodes_at_given_level(self,root,level):
        if self.root==None:
            return 0
        if level==1:
            return 1
        return self.get_nodes_at_given_level(root.left,level-1)+self.get_nodes_at_given_level(root.right,level-1)
```

```python
package binarytree;

class Node {
  Node left;
  Node right;
  int data;
}

class BinaryTree {

  public int getWidthOfLevel(Node node, int level) {
    if (node == null) {
      return 0;
    }

    if (level == 1) {
      return 1;
    }

    return getWidthOfLevel(node.left, level - 1) + getWidthOfLevel(node.right, level - 1);
  }

  public Node createNewNode(int val) {
    Node newNode = new Node();
    newNode.data = val;
    newNode.left = null;
    newNode.right = null;
    return newNode;
  }
}

public class BinaryTreeApp {

  public static void main(String[] args) {

    BinaryTree a = new BinaryTree();

    Node root = a.createNewNode(2);
    root.left = a.createNewNode(7);
    root.right = a.createNewNode(5);
    root.left.left = a.createNewNode(12);
    root.left.right = a.createNewNode(6);
    root.left.right.left = a.createNewNode(5);
    root.left.right.right = a.createNewNode(11);
    root.right.right = a.createNewNode(9);
    root.right.right.left = a.createNewNode(4);
    root.right.right.right = a.createNewNode(40);
    
    System.out.println(a.getWidthOfLevel(root, 3));
  }

}
```
